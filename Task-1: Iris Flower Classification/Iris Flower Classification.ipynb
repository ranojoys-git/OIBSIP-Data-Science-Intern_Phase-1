```python
new=iris.groupby(['Species']).mean()
new
```

```python
plt.subplots(figsize=(8,6))
sns.barplot(new.index,new.SepalLengthCm.values,palette='Blues_d')
plt.title('Sepal Length')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length')
plt.xticks(rotation= 45)
plt.show()
```

```python
plt.subplots(figsize=(8,6))
sns.barplot(new.index,new.SepalWidthCm.values,palette='Blues_d')
plt.title('Sepal Width')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Width')
plt.xticks(rotation= 45)
plt.show()
```

```python
plt.subplots(figsize=(8,6))
sns.barplot(new.index,new.PetalLengthCm.values,palette='Blues_d')
plt.title('Petal Length')
plt.xlabel('Species')
plt.ylabel('Mean Petal Length')
plt.xticks(rotation= 45)
plt.show()
```

```python
plt.subplots(figsize=(8,6))
sns.barplot(new.index,new.PetalWidthCm.values,palette='Blues_d')
plt.title('Petal Width')
plt.xlabel('Species')
plt.ylabel('Mean Petal Width')
plt.xticks(rotation= 45)
plt.show()
```
